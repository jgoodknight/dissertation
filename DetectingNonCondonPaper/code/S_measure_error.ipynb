```python
import math
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def peak_ratio(S, c, k):
    return S  * np.abs( (1 + c * np.sqrt(S) * ( 1 - (k + 1)/S)))**2/np.abs((1 + c * np.sqrt(S) * ( 1 - k/S)))**2 / (k + 1)
```

```python
S_values = [.005, .01, .2, .9]

c_values = np.logspace(-3, 0, 200)
plt.figure()
for s_val in S_values:
    errors = peak_ratio(s_val, c_values, 0) / s_val 
    plt.loglog(c_values, errors, label=r"$S=%.3f$" % s_val)
plt.loglog(c_values, np.ones(c_values.shape), "black")
plt.legend(loc=0)
plt.ylabel(r"$S_{eff} / S $")
plt.xlabel(r"$c$")
plt.title(r"Error in Measuring $S$")
plt.savefig("error_plot_sVerr.png")
```

```python
S_values = np.logspace(-3, 1, 200)
c_values = [.001, .01, .1, .3]
plt.figure()
for c in c_values:
    errors = peak_ratio(S_values, c, 0) / S_values 
    plt.loglog(S_values, errors, label = r"$c=%.3f$" % c)
plt.legend(loc=0)
plt.ylabel(r"$S_{eff} / S $")
plt.xlabel(r"$S$")
plt.title(r"Error in Measuring $S$")
plt.savefig("error_plot_cVerr.png")
```
