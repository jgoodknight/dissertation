```python
import numpy as np
import matplotlib.pyplot as plt
import joeyPlot

%matplotlib inline
```

```python
x = np.linspace(-3, 5)
m = 1
hbar = 1

dx = 1
dE = 1
k_g = 1.0
k_e = .9

s = 1.0
```

```python
v_g = k_g * x**2
v_e = k_e * (x - dx)**2 + dE
psi_g = np.exp(-x**2/(2 *s**2) ) / np.pi**.25
```

```python
plt.figure()
plt.plot(x, v_g)
plt.plot(x, v_e)
plt.plot(x, psi_g)
plt.plot(x, psi_g + dE)
plt.ylim(0, 5)
```
