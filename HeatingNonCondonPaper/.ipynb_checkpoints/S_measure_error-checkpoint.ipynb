```python
import math
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def peak_ratio(S, c, k):
    return S  * ( (1 + c * np.sqrt(S) * ( 1 + (k + 1)/S))/(1 + c * np.sqrt(S) * ( 1 + k/S)))**2 / (k + 1)
```

```python
S_values = [.005, .01, .2, .9]

c_values = np.logspace(-3, 0)
plt.figure()
for s_val in S_values:
    errors = peak_ratio(s_val, c_values, 0) / s_val
    plt.loglog(c_values, errors, label="S=%.3f" % s_val)
plt.legend(loc=0)
plt.ylabel(r"$S_{eff} / S$")
plt.xlabel(r"$c$")
plt.ylim(1, 100)
plt.title(r"Error in Measuring $S$")
```
